### Preparación features de interés

* Voy a agregar un nueva columna, relacionada, precio-m2

* Renombrar columnas (departament y municipio)

```python
#Cambio de nombre
df.rename(columns={"l2":"departamento", "l3":"municipio" }, inplace=True)

#Listo las columnas, para verifico
df.columns
```

* Eliminar columnas con pocos datos

```python
#genero lista con features de interés
features = ["lat", "lon", "municipio", "departamento", "price", "price_m2"]

#Genero lista con features a eliminar
featuresToDel = []
for feature in list(df.columns):
  if feature in features:
    continue
  else:
    featuresToDel.append(feature)
```

```python
# eilimino
df.drop(featuresToDel, axis=1, inplace=True)

#valido
df.columns
```

### Eliminar outliers en price (Por fuera de 3 desviaciones estándar, para dist normal, por fuera aprox de 0.1 y 0.9)

* Previo a esto, guardo los datos totales en un nuevo DataFrame

```python
path_clean = "/content/drive/MyDrive/Colab Notebooks/CienciaDeDatos/01_ML/co_properties_clean.csv"
df.to_csv(path_clean)
```

* Saco los quantiles para obtener los rangos de los precios

```python
price_ranges = list(df.price.quantile([0.1, 0.9]))
price_ranges
```

* Borrar outliers 

```python
# # Creo una mascara/condicional, para hacer filtrado por condicional
mask = df.price.between(price_ranges[0], price_ranges[1])
df = df[mask]
```
