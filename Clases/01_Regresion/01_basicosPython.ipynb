* Lista de diccionarios, json

```python
dataset = []
observaciones = range(0,5)

acum = np.zeros(4)

for element in observaciones:

  peso = np.random.randint(45,80)
  talla = np.random.randint(160, 180) / 100
  calzado = np.random.randint(35,44)

  imc = peso / talla**2

  obs = {
      "peso": peso,
      "talla": talla,
      "calzado": calzado,
      "imc": imc
  }

  acum[0]  += peso
  acum[1]  += talla
  acum[2]  += calzado
  acum[3]  += imc

  dataset.append(obs)

print(dataset)
print(acum)

media = []
for element in acum:
  media_elem = element / len(dataset)

  media.append(media_elem)
  

print(media)
```

# Operaciones por columnas (diccionario d listas)

```python
data = {}

observaciones = range(0,5)
media = 0
  
peso = np.random.randint(45,80, 3)
talla = np.random.randint(160, 180, 3) / 100
calzado = np.random.randint(35,44,3 )

imc = peso / talla**2

data["masa"] = list(peso)
data["talla"] = list(talla)
data["calzado"] = list(calzado)
data["imc"] = list(imc)

data
```
